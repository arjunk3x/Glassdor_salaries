```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/roboc/Downloads/glassdoor_jobs.csv")
```

```python
df=df.loc[df['Salary Estimate']!="-1"]
```

```python
df['Salary Estimate'] = df['Salary Estimate'].str.replace('(Glassdoor est.)', '')
```

```python
df['Salary Estimate'].unique()
```

```python
df['Salary Estimate'].unique()
```

```python
df['Salary Estimate'] = df['Salary Estimate'].str.replace('(Employer est.)', '')
```

```python
df['Salary Estimate'] = df['Salary Estimate'].str.replace('K', '')
df['Salary Estimate'] = df['Salary Estimate'].str.replace('$', '')
```

```python
df['Salary Estimate'].unique()
```

```python
df
```

```python
def fun(x):
    if "Employer Provided Salary" in x:
        return 1
    else:
        return 0

df['Employer_Provided_Salary']=df['Salary Estimate'].apply(lambda x:fun(x))
```

```python
def funn(x):
    if "Per Hour" in x:
        return 1
    else:
        return 0

df['Per Hour']=df['Salary Estimate'].apply(lambda x:funn(x))
```

```python
df
```

```python
df['Salary Estimate'] = df['Salary Estimate'].str.replace(' Per Hour', '')
df['Salary Estimate'] = df['Salary Estimate'].str.replace('Employer Provided Salary:', '')
```

```python
df['Salary Estimate'].unique()
```

```python
def extract_min_salary(x):
    return int(x.split('-')[0])

def extract_max_salary(x):
    return int(x.split('-')[1])

df['min_salary'] = df['Salary Estimate'].apply(extract_min_salary)
df['max_salary'] = df['Salary Estimate'].apply(extract_max_salary)
```

```python
def cmp(x):
    for i in range(len(x)-1):
        if x[i] == "\n":
            x = x[0:i]
            break
    return x

df['Company_txt'] = df['Company Name'].apply(cmp)
```

```python
def loc_change(x):
    result_list = x.split(',')
    return ''.join(result_list[1:3]).upper()

df['Location_abb'] = df['Location'].apply(lambda x: loc_change(x))
```

```python
df['Location_abb'].value_counts()
```

```python
df
```

```python
def loc_same(row):
    if row['Location'] == row['Headquarters']:
        return 1
    else:
        return 0

df['Location_same'] = df.apply(lambda row: loc_same(row), axis=1)
```

```python
df['Location_same']
```

```python
df['avg_salary']=(df['max_salary']+df['min_salary'])/2
```

```python
df['avg_salary']
```

```python
df['Founded'].unique()
```

```python
def since(x):
    if x<1:
        return x
    else:
        return (2020-x)

df['Since_Years'] = df['Founded'].apply(lambda x: since(x))
```

```python
df['Since_Years']
```

```python
#Python
df['python_yn'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
```

```python
#R
df['R_yn'] = df['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() or 'r-studio' in x.lower() else 0)
```

```python
#Spark
df['spark'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
```

```python
#aws
df['aws'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
csv_file_path = 'C:/Users/roboc/Downloads/salaries_cleaned.csv'
df.to_csv(csv_file_path, index=False)
```
